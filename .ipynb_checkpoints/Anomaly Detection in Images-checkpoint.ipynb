From images like the one above, a dataset has been created consisting of **28739** 64x64 pixel images, figure (a) below. Of all these images, only **61** of them has been manually anotated as *bad* images (containing at least one defect) and the rest has been annotated as *good* images (no defects present). Example of defects can been seen in the images below.

![dataOverview](figures/dataOverview.jpg)

## The problem

We are dealing with a binary classification problem. Our goal is to predict if a given image contains a defect circuit segment or not. Since the dataset is very imbalanced, we cannot rely on methods like Logistic Regression, Support Vector Machines or even Neural Network based classifier models. 

To solve this, we will use an *Autoencoder*. The Autoencoder takes an image as input, compresses it through a *encoder* network into a low dimensional representation (bottleneck layer), and then expand the data back to its original size through a *decoder* network. The output will therefore also be an image in this case. 

How can we use this for anomaly detection?

The Autoencoder is trained to reconstruct the input image from the bottleneck layer as good as possible. This performance can be measured by a chosen *reconstruction error*, which rely on some difference between the input and the output of the model. We can utilize this for anomaly detection by having the model solely trained on the samples from the major class (good images in our case). When feeding a bad image to the trained model, we can then expect to see a much higher reconstruction error than for the good images, since this is not what the model has been trained to do a good reconstruction of. If the model is trained properly, it would therefore be possible to separate the classes through a chosen threshold on the reconstruction error.

Let's try it!

## Loading the data

We'll read the images in gray scale and reshape them to fit Tensorflow's framework.

```python
import numpy as np
from pathlib import Path
import os
import cv2
from tqdm import tqdm

data_dir = 'circuit_images'
img_size = 64
categories = ["good", "bad"]
X = []
y = []

for i, category in enumerate(categories):
    print('Loading images in category:', category, ', value', i)

    for (dirpath, dirnames, images) in os.walk(os.path.join(data_dir, category)):
        for img in tqdm(images):
            try:
                img_array = cv2.imread(os.path.join(dirpath, img), cv2.IMREAD_GRAYSCALE)
                if img_array is None:
                    raise IOError('\n Cannot load: ' + os.path.join(dirpath, img))

                resized_img_array = cv2.resize(img_array, (img_size, img_size))
                X.append(resized_img_array)
                y.append(i)

            except IOError as e:
                print(str(e))
                pass

# Reshape data to agree with Tensorflow
images = np.array(X).reshape(-1, img_size, img_size, 1)
labels = y
```

## The model

We will now construct the Autoencoder. The Encoder part will consist of a series of downsampling layers Conv2D+MaxPooling2D, down to the bottleneck layer with dimension (512,4,4). The Decoder part consist of a series of Conv2D+UpSampling2D layers to upsample the data.

```python
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D
from tensorflow.keras.models import Model

input_img = Input(shape=(img_size, img_size, 1))
x = Conv2D(64, (3, 3), activation='relu', padding='same')(input_img)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(128, (3, 3), activation='relu', padding='same')(x)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(256, (3, 3), activation='relu', padding='same')(x)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(512, (3, 3), activation='relu', padding='same')(x)
# The latent layer
encoded = MaxPooling2D((2, 2), padding='same')(x)

x = Conv2D(512, (3, 3), activation='relu', padding='same')(encoded)
x = UpSampling2D((2, 2))(x)
x = Conv2D(256, (3, 3), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)
x = Conv2D(128, (3, 3), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)
x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)
decoded = Conv2D(1, (3, 3), activation='sigmoid', padding='same')(x)
autoencoder = Model(input_img, decoded)

# Create the Autoencoder model
autoencoder.compile(optimizer='adam',
                    loss='mse')
autoencoder.summary()
```

## Normalizing and train

Before training, the image pixel values need to be normalized. Since they are RGB values, the maximum pixel value is 255, and we therefore normalize by dividing the data by 255.

```python
images = images.astype('float32') / 255.
```

For now, we only want to train on good images. Let's extract the good images from the data and use 80 % of them for training the Autoencoder.

```python
good_images = np.array([images[i] for i, j in enumerate(labels) if j == 0])
bad_images = np.array([images[i] for i, j in enumerate(labels) if j == 1])

x_train, x_test = np.split(good_images, [int(0.8 * len(good_images))])
```

```python
# Fit the model
history = autoencoder.fit(x_train, x_train,
                          epochs=60,
                          batch_size=32,
                          shuffle=True,
                          validation_split=0.3,
                          )
```

This training process can take a while on a normal CPU. For now, let's load a model that's been trained at an earlier moment.

## Evaluate the model

```python
from tensorflow import keras

autoencoder = keras.models.load_model("CAE_model.h5")

# Select test images
good_test = x_test
bad_test = bad_images

# Predict on the image, i.e. generate the reconstructed images
good_reconst = autoencoder.predict(good_test)
bad_reconst = autoencoder.predict(bad_test)
```

### The Reconstruction Error

One way to measure the error between the input and the reconstructed image is simply to take the mean squared error (MSE) between the pixel values over the whole image, i.e.

$$ MSE = \frac{1}{N}\sum_{i=1}^N (x_{i}-y_{i})^2$$

During this project, we found that calculating the MSE over sections of $p \times p$ pixels, and then choosing the maximum MSE value from these sections as our metric, performed significantly better in separating the good images from the bad images. We call this metric MMSE*p* and it can be explained as

$$ MMSEp = \max_j \frac{1}{p^2}\sum_{i=1}^{p^2} (x^j_i-y^j_i)^2 $$ 

where the index $j$ represents all the $p \times p $ sections in the image. By means of experiment, $p=8$ seemed to be the best choice for this problem, which is what we'll use here.

```python
# Calculates the reconstruction error MMSEp for given p, input and output image.
# Returns MMSEp and the location for the error
def max_sectional_sqe(img, decoded_img, window_dim):
    if window_dim > len(img[0]):
        window_dim = int(np.sqrt(img[0]))

    max_arr_size = len(img[0]) - window_dim + 1
    max_arr = np.zeros((max_arr_size, max_arr_size))
    img_sqe = img - decoded_img
    img_sqe = np.multiply(img_sqe, img_sqe) / (window_dim * window_dim)

    for i in range(max_arr_size):
        for j in range(max_arr_size):
            max_arr[i][j] = (np.sum(img_sqe[i: i + window_dim, j: j + window_dim]))

    max_sqe = np.max(max_arr)
    ind = np.unravel_index(np.argmax(max_arr, axis=None), max_arr.shape)
    ind = [ind[1], ind[0]]
    return [max_sqe, ind]
```

### Visualization
We'll now use the above function max_sectional_sqe to calculate the reconstruction error for a few selected good and bad images.


```python
import matplotlib.pyplot as plt
import matplotlib.patches as pat

n = 8
window_size = 8
k_g = 126
k_b = 21
fig_good = plt.figure(1, figsize=(15, 4))
fig_bad = plt.figure(2, figsize=(15, 4))

for i in range(1, n):

    sqe_good = max_sectional_sqe(good_test[i+k_g], good_reconst[i+k_g], window_size)
    sqe_bad = max_sectional_sqe(bad_test[i+k_b], bad_reconst[i+k_b], window_size)

    # Figure for good images
    plt.figure(1)

    # Original
    ax1 = plt.subplot(2, n, i)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(good_test[i+k_g].reshape(img_size, img_size))
    plt.gray()
    rect = pat.Rectangle((sqe_good[1]), window_size, window_size, edgecolor='r', facecolor="none")
    plt.gca().add_patch(rect)

    # Reconstructed
    plt.subplot(2, n, i + n)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(good_reconst[i+k_g].reshape(img_size, img_size))
    plt.gray()
    ax = plt.gca()
    label = ax.set_xlabel(round(sqe_good[0], 3), fontsize=12)
    ax.xaxis.set_label_coords(0.5, -0.15)

    # Figure for bad images
    plt.figure(2)

    # Original
    ax2 = plt.subplot(2, n, i)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(bad_test[i+k_b].reshape(img_size, img_size))
    plt.gray()
    rect = pat.Rectangle((sqe_bad[1]), window_size, window_size, edgecolor='r', facecolor="none")
    plt.gca().add_patch(rect)

    # Reconstructed
    plt.subplot(2, n, i + n)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(bad_reconst[i+k_b].reshape(img_size, img_size))
    plt.gray()
    ax = plt.gca()
    label = ax.set_xlabel(round(sqe_bad[0], 3), fontsize=12)
    ax.xaxis.set_label_coords(0.5, -0.15)

ax1.text(-300, -15, 'Reconstruction of non-defect circuit images', fontsize=16, family='serif')
ax1.text(-510, 35, 'Original', fontsize=12, family='serif')
ax1.text(-540, 130, 'Reconstructed', fontsize=12, family='serif')
ax1.text(-510, 174, 'MMSE8', fontsize=12, family='serif')

ax2.text(-300, -15, 'Reconstruction of defect circuit images', fontsize=16, family='serif')
ax2.text(-510, 35, 'Original', fontsize=12, family='serif')
ax2.text(-540, 130, 'Reconstructed', fontsize=12, family='serif')
ax2.text(-510, 174, 'MMSE8', fontsize=12, family='serif')

plt.show()
```

From the images above we can se the the Autoencoder basically removes the noise in the images. For the defect circuit images, the defect part of the image is sort of replaced with a non-defect segment in the reconstructed image. We can also see that the error MMSE8 is substantially higher for the defect circuit images. This is what we wanted to achieve.

Let's calculate MMSE8 for all the test images and visualize it in a scatter plot.

```python
sqe_bad = [max_sectional_sqe(bad_test[i], bad_reconst[i], window_size)[0] for i in tqdm(range(len(bad_test)))]
sqe_good = [max_sectional_sqe(good_test[i], good_reconst[i], window_size)[0] for i in tqdm(range(int(0.2 * len(good_test))))]

fig, ax1 = plt.subplots(figsize=(15,6))
ax2 = ax1.twiny()

x_bad = range(len(sqe_bad))
x_good = range(len(sqe_good))

scatter1 = ax1.scatter(x_good, sqe_good, label="Non-defect", color='0.6', marker='o')
scatter2 = ax2.scatter(x_bad, sqe_bad, label="Defect", color='0.1', marker='D')


scatters = [scatter1, scatter2]

ax2.legend(scatters, [scatter.get_label() for scatter in scatters])

ax1.set_xlabel("Non-defect test images", fontsize=14, family='serif')
ax2.set_xlabel("Defect test images", fontsize=14, family='serif')

ax1.set_ylabel("MMSE8", fontsize=14, family='serif')

ax1.tick_params(axis='y')

ax1.tick_params(axis='x')
ax2.tick_params(axis='x')

plt.title("Reconstruction error MMSE8 for defect and non-defect test images", fontsize=16, family='serif', pad=30)
plt.subplots_adjust(top=0.83,
                    bottom=0.07,
                    left=0.165,
                    right=0.82,
                    hspace=0.2,
                    wspace=0.2)
plt.show()
```

The above figure makes it clear that the defect circuit images can be quite easy linearly separated from the non-defect circuit images, by means of the reconstruction error MMSE8.

## Use of model
To further use this model, we would select a threshold $T$ for the reconstruction error MMSE8, and classify new images with MMSE8 > $T$ as defect and non-defect otherwise.

## Conclusion and remarks
This Autoencoder can successfully be used to classify an extremely imbalanced 2-class dataset, i.e. separate the rare samples from the numerous ones. Worth mentioning is that the bad images wouldn't have to be excluded from the training process. Since they are just a fraction of the good images, they would have negligeable impact on the final weights and biases for the model. This means that the problem can be solved completely *unsupervised*, meaning that we don't even have to annotate the images to begin with. The approach could just be to train on all images, and then manually go through the ones with the highest MMSE8 until you've reached images that doesn't display any anomaly.
